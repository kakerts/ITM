```python
import requests
import time
import matplotlib.pyplot as plt

latitude = []  
#list to store latitude data points
longitude = []  
#list to store longitude data points

for _ in range(100):
#for loop runs 100 times to generate at least 100 data points
    response = requests.get('http://api.open-notify.org/iss-now.json')
    data = response.json()
    latitude.append(float(data['iss_position']['latitude']))  
    #get latitude and convert to float and append to list
    longitude.append(float(data['iss_position']['longitude'])) 
    #get longitude and convert to float and append to list
    time.sleep(10)  
    #waits for 10 seconds before generating next data point

#usematplotlib to illustrate path of ISS

plt.plot(longitude, latitude, 'o-')  
#plot latitude vs longitude with marker at each data point
plt.xlabel('Longitude')  
#set x-axis label
plt.ylabel('Latitude')  
#set y-axis label
plt.title('Path of the ISS')  
#set title of plot
plt.grid(True) 
#add gridlines to plot
plt.show()  
#display plot
```
