1. How many unique customers are in the dataset?

2. What are the different categories of products available? How many unique categories are there?

```python
import pandas as pd
df = pd.read_csv('Downloads/fct_invoice.csv')

categories = df['category'].unique()

num_categories = df['category'].nunique()

print(categories)
print(num_categories)
```

3. Which payment method is the most popular? How many times was it used?

```python
import pandas as pd

df = pd.read_csv('Downloads/fct_invoice.csv')
payment_counts = df['payment_method'].value_counts()
most_popular_payment = payment_counts.idxmax()
count_most_popular_payment = payment_counts.max()

print(most_popular_payment)
print(count_most_popular_payment)
```

1. What are the three most popular categories, by total sales?

```python
import pandas as pd

df = pd.read_csv('Downloads/fct_invoice.csv')
category_sales = df.groupby('category')['quantity'].sum()
sorted_sales = category_sales.sort_values(ascending=False)
top_categories = sorted_sales.head(3)

print(top_categories.index.tolist())
```

2. What are the total sales attributed to customers over the age of 45?

```python
import pandas as pd

age_df = pd.read_json('Downloads/dim_customer.json')
sales_df = pd.read_csv('Downloads/fct_invoice.csv')

merged_df = pd.merge(age_df, sales_df, left_on='id', right_on='customer_id')
#merge dataframes
filtered_df = merged_df[merged_df['age'] > 45]
total_sales = filtered_df['price'].sum()

print(total_sales)
```

3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
import pandas as pd
import matplotlib.pyplot as plt

invoices_df = pd.read_csv('Downloads/fct_invoice.csv')

invoices_df['invoice_date'] = pd.to_datetime(invoices_df['invoice_date'])
#convert the 'invoice_date' column to datetime format
invoices_df['month_year'] = invoices_df['invoice_date'].dt.to_period('M')
#extract month and year from 'invoice_date' column
monthly_counts = invoices_df['month_year'].value_counts().sort_index()
#count number of invoices for each month

plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='line', marker='o')
plt.title('Distribution of Invoices Over Time')
plt.xlabel('Month and Year')
plt.ylabel('Number of Invoices')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).

```python
import pandas as pd

age_df = pd.read_json('Downloads/dim_customer.json')
sales_df = pd.read_csv('Downloads/fct_invoice.csv')

merged_df = pd.merge(age_df, sales_df, left_on='id', right_on='customer_id')
#merge age and sales data based on customer_id column
merged_df['decade_age_range'] = pd.cut(merged_df['age'], bins=range(0, 101, 10), right=False)
#create new column for decade age range
pivot_table = pd.pivot_table(merged_df, values='price', index='category', columns='decade_age_range', aggfunc='sum')
#pivot table showing breakdown of sales by category and decade age range

print(pivot_table)
```
